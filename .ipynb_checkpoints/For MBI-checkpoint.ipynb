```python
import pdfquery
pdf = pdfquery.PDFQuery("P35_ALBANESE_Anthony.pdf")
pdf.load()
tree = pdf.get_tree()
tree.write("albo.xml", pretty_print=True)
```

```python
for element in tree.iter():
    if element.tag == 'LTTextLineHorizontal' or element.tag == 'LTTextBoxHorizontal':
        if element.text.strip() != '':
            print element.text, element.get('height')
```

```python
import pdfquery
pdf = pdfquery.PDFQuery("P35_BRODTMANN_Gai.pdf")
pdf.load()
tree = pdf.get_tree()
```

```python

for element in tree.iter():
    if element.tag == 'LTTextLineHorizontal' or element.tag == 'LTTextBoxHorizontal':
        if element.text.strip() != '':
            if float(element.get('height')) > 17:
                print element.text, float(element.get('height'))
                for part in element.iterancestors(tag="LTPage"):
                    print part.get("page_index")
```
