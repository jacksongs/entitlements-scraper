```python
# Let's get every major heading, based on its size, and save it do a dataframe.
localpdfsbase = !cd pdfs;ls # Gets the filenames


data = []

for pdffile in localpdfsbase[:10]:

    pdf = pdfquery.PDFQuery("pdfs/"+pdffile)
    pdf.load()
    tree = pdf.get_tree()
    for element in tree.iter():
        if element.tag == 'LTTextLineHorizontal' or element.tag == 'LTTextBoxHorizontal':
            if element.text.strip() != '':
                if float(element.get('height')) > 17:
                    row = {}
                    row['text'] = element.text.strip()
                    row['size'] = float(element.get('height'))
                    for part in element.iterancestors(tag="LTPage"):
                        row['page'] = part.get("page_index")
                    row['file'] = pdffile
                    data.append(row)
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```
